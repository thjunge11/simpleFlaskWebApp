# Backup database on the host:
```bash
docker run --rm --volumes-from simpleflaskwebapp-db-1 -v $(pwd):/backup ubuntu tar cvf /backup/gaming_postgresDB.tar /var/lib/postgresql/data
```

# Restore: 
```bash
docker run --rm --volumes-from simpleflaskwebapp-db-1 -v $(pwd):/backup ubuntu bash -c "cd /var/lib/postgresql/data && tar xvf /backup/gaming_postgresDB.tar --strip 1"
```
